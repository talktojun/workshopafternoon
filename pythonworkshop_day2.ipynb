```python
% matplotlib inline
avg_inflammation = data.mean(axis=0)
avg_plot = matplotlib.pyplot.plot(avg_inflammation)

matplotlib.pyplot.xlabel("Days")
matplotlib.pyplot.ylabel("Values")
matplotlib.pyplot.title("My purty plot")

matplotlib.pyplot.show(avg_plot)
```

```python

fig = matplotlib.pyplot.figure(figsize = (10.0, 3.0))

axes1 = fig.add_subplot(1, 3, 1)
axes2 = fig.add_subplot(1, 3, 2)
axes3 = fig.add_subplot(1, 3, 3)

axes1.set_ylabel("Max")
axes1.plot(data.max(axis=0))

axes2.set_ylabel("Min")
axes2.plot(data.min(axis=0))

axes3.set_ylabel("Mean")
axes3.plot(data.mean(axis=0))

fig.tight_layout()

```
